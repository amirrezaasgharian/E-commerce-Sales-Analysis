# <p style="padding:10px;background-color:#00bfff ;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">Total Sales by category + (Min and Max)</p>

```python
#Total sales across the year
df['total_sales'] = df.iloc[:, 6:18].sum(axis=1)
category_sales = df.groupby('category')['total_sales'].sum().reset_index()
monthly_sales = df.groupby('category').sum().iloc[:, 5:17].reset_index()
```

```python
max_sales_value = category_sales['total_sales'].max()
min_sales_value = category_sales['total_sales'].min()

max_sales_category = category_sales.loc[category_sales['total_sales'] == max_sales_value, 'category'].values[0]
min_sales_category = category_sales.loc[category_sales['total_sales'] == min_sales_value, 'category'].values[0]

plt.figure(figsize=(10, 6))
bars = plt.bar(category_sales['category'], category_sales['total_sales'], color='skyblue')

for bar in bars:
    height = bar.get_height()
    if height == max_sales_value:
        plt.text(bar.get_x() + bar.get_width() / 2, height, 'Max', ha='center', va='bottom', fontsize=10, color='green')
    elif height == min_sales_value:
        plt.text(bar.get_x() + bar.get_width() / 2, height, 'Min', ha='center', va='bottom', fontsize=10, color='red')

plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<a id="4"></a>
# <p style="padding:10px;background-color:#00bfff ;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">Monthly Sales Trend by Category</p>

```python
plt.figure(figsize=(14, 8))
for category in monthly_sales['category']:
    plt.plot(range(1, 13), monthly_sales[monthly_sales['category'] == category].iloc[0, 1:], label=category)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend by Category')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<a id="5"></a>
# <p style="padding:10px;background-color:#00bfff ;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">Total Monthly Sales + (Min and Max)</p>

```python
total_monthly_sales = df.iloc[:, 6:18].sum()

max_sales_month = total_monthly_sales.idxmax()
max_sales_value = total_monthly_sales.max()

min_sales_month = total_monthly_sales.idxmin()
min_sales_value = total_monthly_sales.min()

mean_sales = total_monthly_sales.mean()

print(f"Max Sales Month: {max_sales_month} with Total sales of: {max_sales_value}")
print(f"Min Sales Month: {min_sales_month} with Total sales of: {min_sales_value}")
```

```python
plt.figure(figsize=(14, 8))
plt.plot(range(1, 13), total_monthly_sales, marker='o')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Monthly Sales')
plt.xticks(range(1, 13), [f'Month {i}' for i in range(1, 13)])
plt.grid(True)

plt.text(int(max_sales_month.split('_')[-1]), max_sales_value, 'Max', ha='center', va='bottom', fontsize=16, color='green')
plt.text(int(min_sales_month.split('_')[-1]), min_sales_value, 'Min', ha='right', va='bottom', fontsize=16, color='red')

plt.axhline(mean_sales, color='orange', linestyle='--')

plt.tight_layout()
plt.show()
```

<a id="6"></a>
# <p style="padding:10px;background-color:#00bfff ;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">Review Analysis</p>

```python
#Best and worth category based on the total sales and review score
category_review_scores = df.groupby('category')['review_score'].mean()
```

```python
best_review_category = category_review_scores.idxmax()
worst_review_category = category_review_scores.idxmin()
```

```python
print(f"The Best category: {best_review_category} with mean score of: {category_review_scores.max()}")
print(f"The Worst category: {worst_review_category} with mean score of: {category_review_scores.min()}")
```

```python
plt.figure(figsize=(14, 8))
bars = plt.bar(category_review_scores.index, category_review_scores.values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Average Review Score')
plt.title('Average Review Score by Category')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    if height == category_review_scores.max():
        plt.text(bar.get_x() + bar.get_width() / 2, height, 'Best', ha='center', va='bottom', fontsize=10, color='green')
    elif height == category_review_scores.min():
        plt.text(bar.get_x() + bar.get_width() / 2, height, 'Worst', ha='center', va='bottom', fontsize=10, color='red')

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Most favorite product and category based on the review_count and review_score

df['weighted_review_score'] = df['review_score'] * df['review_count']

most_favorite_product = df.loc[df['weighted_review_score'].idxmax()]

most_reviewed_product = df.loc[df['review_count'].idxmax()]
```

```python
category_stats = df.groupby('category').agg(
    average_review_score=('review_score', 'mean'),
    total_review_count=('review_count', 'sum')
)
```

```python
most_favorite_category = category_stats.loc[category_stats['average_review_score'].idxmax()]
```

```python
print(f"Most Favorite Product based on Weighted Review Score: {most_favorite_product['product_name']} in {most_favorite_product['category']}")
print(f"Most Reviewed Product based on Review Count: {most_reviewed_product['product_name']} in {most_reviewed_product['category']}")
print(f"Most Favorite Category based on Average Review Score: {most_favorite_category.name}")
```

# <p style="padding:10px;background-color:#00bfff ;margin:0;color:#ffffff;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">Best and Worst product of each Category</p>

```python
#Top and bottom Product of each category based on the total sales
grouped = df.groupby('category')

def get_top_and_bottom_product_names(group):
    sorted_group = group.sort_values(by='total_sales', ascending=False)
    top = sorted_group.head(1)[['product_name', 'total_sales']].values[0]
    bottom = sorted_group.tail(1)[['product_name', 'total_sales']].values[0]
    return pd.Series({'top': top[0], 'top_sales': top[1], 'bottom': bottom[0], 'bottom_sales': bottom[1]})

result = grouped.apply(get_top_and_bottom_product_names).reset_index()
```

```python
result
```

```python
figs = []
for index, row_data in result.iterrows():
    top_product_sales = row_data['top_sales']
    total_sales = grouped.get_group(row_data['category'])['total_sales'].sum()
    other_sales = total_sales - top_product_sales

    labels = [f"Top Product: {row_data['top']} ({top_product_sales})", f"Other Products ({other_sales})"]
    values = [top_product_sales, other_sales]

    fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent', hole=0.3)])
    fig.update_layout(title=f"Top Product Sales Percentage for {row_data['category']}", showlegend=False)
    figs.append(fig)

for fig in figs:
    fig.show()
```
